```python
annFile = 'annotations/person_keypoints_train2014.json'
coco_kps=COCO(annFile)
```

```python
def show_img_annotations(coco_kps, im_id):
    if(not im_id):
        imgIds = coco_kps.getImgIds(catIds=catIds);
        im_id = imgIds[np.random.randint(0,len(imgIds))]
        
    img = coco_kps.loadImgs(im_id)[0]
    I = io.imread(img['coco_url'])
    plt.imshow(I); plt.axis('off')
    ax = plt.gca()
    annIds = coco_kps.getAnnIds(imgIds=img['id'], catIds=[1], iscrowd=None)
    anns = coco_kps.loadAnns(annIds)

    for item in anns:
        if 'segmentation' in item:
            del item['segmentation']
    
    coco_kps.showAnns(anns)    
```

```python
show_img_annotations(coco_kps, 264551)
```
