```python
%%time
res = net(ds[0][0][None])
```

```python
net.train()
n_epochs = 3
base_lr = 0.0005
#4e-5
optimizer = optim.Adam([
        {'params': get_parameters_conv(net.model, 'weight')},
        {'params': get_parameters_conv_depthwise(net.model, 'weight'), 'weight_decay': 0},
        {'params': get_parameters_bn(net.model, 'weight'), 'weight_decay': 0},
        {'params': get_parameters_bn(net.model, 'bias'), 'lr': base_lr * 2, 'weight_decay': 0},
        {'params': get_parameters_conv(net.initial_stage, 'weight'), 'lr': base_lr},
        {'params': get_parameters_conv(net.initial_stage, 'bias'), 'lr': base_lr * 2, 'weight_decay': 0},
        {'params': get_parameters_conv(net.refinement_stages, 'weight'), 'lr': base_lr * 4},
        {'params': get_parameters_conv(net.refinement_stages, 'bias'), 'lr': base_lr * 8, 'weight_decay': 0},
        {'params': get_parameters_bn(net.refinement_stages, 'weight'), 'weight_decay': 0},
        {'params': get_parameters_bn(net.refinement_stages, 'bias'), 'lr': base_lr * 2, 'weight_decay': 0},
    ], lr=base_lr, weight_decay=5e-4)


drop_after_epoch = [50, 100, 150]
scheduler = optim.lr_scheduler.MultiStepLR(optimizer, milestones=drop_after_epoch, gamma=0.333)


#optimizer = torch.optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr = lr)

total_steps = 0

for e in range(n_epochs):
    for imgs, pafs, paf_inds, hms, hm_inds in dl:
        optimizer.zero_grad()
        
        imgs, pafs, paf_inds, hms, hm_inds = imgs.to(device), pafs.to(device), paf_inds.to(device), hms.to(device), hm_inds.to(device) 
        res = net(imgs)
        loss = model_utils.paf_and_heatmap_loss(res, pafs, paf_inds, hms, hm_inds)
        
        loss.backward()
        optimizer.step()
        
        total_steps +=1
        if(total_steps%250==0):
            torch.save(model.state_dict(),'model-wts.ckpt')
            print('saving model')
                                                       
        model_utils.print_training_loss_summary(loss.item(), total_steps, e+1, n_epochs, len(dl))
```

```python
def make_pose_predictions_and_save_img(im_path, model, IM_SIZE=IM_SIZE, show_plots=False):
    test_tensor_tfms = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize(torch.tensor([0.485, 0.456, 0.406]), torch.tensor([0.229, 0.224, 0.225]))
    ])
    
    Resize_im = Resize(IM_SIZE)
    
    IM_STG_INPUT_SIZE = 46 
    if(IM_SIZE==184):
        IM_STG_INPUT_SIZE = 23    
    
    pf, hm, un_im, pad = model_utils.get_pafs_and_hms_predictions(im_path, model, test_tensor_tfms, Resize_im, IM_SIZE, IM_STG_INPUT_SIZE)     
    joint_pos_map = model_utils.get_joint_positions(hm.cpu().numpy(), nms_window=int(368*30/IM_SIZE))
    part_matches_map = model_utils.calculate_part_matches_from_predicted_joints_and_pafs(joint_pos_map, pf.cpu())
    
    oriImg = cv2.imread(im_path)
    canvas = model_utils.draw_bodypose(oriImg, part_matches_map, pad, IM_SIZE)
    cv2.imwrite(im_path.split('/')[-1], canvas)
    
    if(show_plots):
        plot_utils.plot_pafs(un_im, pf.cpu().numpy(), figsize=(20,20))
        plot_utils.plot_heatmaps(un_im, hm.cpu().numpy(), figsize=(20,20))
        plt.imshow(canvas[:, :, [2, 1, 0]])
        plt.show()
        plot_utils.plot_matching_results(un_im, part_matches_map, joint_pos_map, pf)
```

```python
#1520
from CONSTANTS import SKELETON
def plot_kps(ann):
    plt.plot(ann[:,:,0], ann[:,:,1], 'ro')
    
def plot_kps_pairs(anns):
    for person in anns:
        for idx1, idx2 in SKELETON:
            pts = person[np.array([idx1, idx2])]
            if((person[idx1][0]>0 and person[idx1][1]>0) and (person[idx2][0]>0 and person[idx2][1]>0)):
                plt.plot(pts[:,0], pts[:,1], lw=2, color='w')
            
def plot_ds_image(img, kps):
    plt.imshow(img)
    plot_kps(kps)
    plot_kps_pairs(kps)
```

```python
from PIL import Image

def plot_points(ann, c=None):
    colors = ['r', 'g', 'b', 'y', 'c']
    if(not c):
        rand_idx = np.random.randint(0, len(colors))
        c = colors[rand_idx]
        
    plt.plot(ann[used_kps][:,0], ann[used_kps][:,1], c+'o')
    
def plot_pairs(ann):
    for pt1, pt2 in part_pairs:
        idx1, idx2 = keypoint_type_to_idx[pt1], keypoint_type_to_idx[pt2] 
        pts = ann[np.array([idx1, idx2])]
        if((ann[idx1][0]>0 and ann[idx1][1]>0) and (ann[idx2][0]>0 and ann[idx2][1]>0)):
            plt.plot(pts[:,0], pts[:,1], lw=2, color='w')

def draw_bbox(center, w, h):
    crop_size = 368
    x,y = center
    if(w>h):
        y = h/2
        crop_size = h
    else:
        x = w/2
        crop_size = w
        
    x_l, x_r = x - crop_size//2, x + crop_size//2
    y_t, y_d = y - crop_size//2, y + crop_size//2
    if(x_l<0):
        x_l = 0
    elif(x_r>w):
        x_r = w
    if(y_t<0):
        y_t = 0
    elif(y_d>h):
        y_d = h
    
    print(x_l, y_t, x_r, y_d)
    tl = np.array([x_l, y_t])
    tr = np.array([x_r, y_t])
    bl = np.array([x_l, y_d])
    br = np.array([x_r, y_d])
    tl_tr = np.array([tl, tr])
    tl_bl = np.array([tl, bl])
    tr_br = np.array([tr, br])
    bl_br = np.array([bl, br])
    plt.plot(tl_tr[:,0], tl_tr[:,1], lw=2, color='black')
    plt.plot(tl_bl[:,0], tl_bl[:,1], lw=2, color='black')
    plt.plot(tr_br[:,0], tr_br[:,1], lw=2, color='black')
    plt.plot(bl_br[:,0], bl_br[:,1], lw=2, color='black')
    
def plot_image_with_skeleton(idx):
    colors = ['r', 'g', 'b', 'y', 'c']
    ann = annotations_data[idx]
    img = Image.open('./images/'+ann['img_paths'])
    plt.imshow(img)
    plot_points(ann['joint_self'], 'r')
    plot_pairs(ann['joint_self'])
    
    if(ann['numOtherPeople']>0):
        for othrs in ann['joint_others']:
            plot_points(othrs)
            plot_pairs(othrs)
    draw_bbox(ann['objpos'], ann['img_width'], ann['img_height'])        
```
